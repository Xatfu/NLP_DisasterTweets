```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train = pd.read_csv('/kaggle/NLP_DisasterTweets/train.csv')
df_test = pd.read_csv('/kaggle/NLP_DisasterTweets/test.csv')
df_sub = pd.read_csv('/kaggle/NLP_DisasterTweets/sample_submission.csv')
```

```python
df_train
```

```python
df_test
```

```python
X = df_train['text']
y = df_train['target']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tfidf = TfidfVectorizer()
```

```python
model = DecisionTreeClassifier()
```

```python
pipe = Pipeline([('tfidf', tfidf), ('model', model)])
```

```python
params = {
    'tfidf__max_features': [1000, 2000, 5000],
    'tfidf__ngram_range': [(1, 1), (1, 2)],
    'model__max_depth': [5, 10, 20, None],
    'model__min_samples_split': [2, 5, 10]
}
```

```python
X_test = df_test['text']
```

```python
final_model = GridSearchCV(pipe, params, cv=3, verbose=2)
```

```python
final_model.fit(X, y)
```

```python
y_pred = final_model.predict(X_test)
```

```python
df_sub['target'] = y_pred
```

```python
df_sub.to_csv('sub.csv', index=False, index_label=False)
```

**WITH RANDOM FOREST**

```python
X = df_train['text']
y = df_train['target']
```

```python
X_test = df_test['text']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
```

```python
tfidf = TfidfVectorizer()
```

```python
model = RandomForestClassifier()
```

```python
pipe = Pipeline([('tfidf', tfidf), ('model', model)])
```

```python
params = {
    'tfidf__max_features': [1000, 3000, 5000],
    'tfidf__ngram_range': [(1, 1), (1, 2), (1, 3)],
    'model__n_estimators': [100, 200, 300],
    'model__max_depth': [10, 20, 30, None],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4]
}
```

```python
final_model = GridSearchCV(pipe, params, cv=3, verbose=2)
```

```python
final_model.fit(X, y)
```

```python
y_pred = final_model.predict(X_test)
```

```python
df_sub['target'] = y_pred
```

```python
df_sub.to_csv('final_sub.csv', index=False, index_label=False)
```
